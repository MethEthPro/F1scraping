```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df_2021_year = pd.read_csv("race.csv",encoding = "unicode_escape" )
df_quali_21_22_abudabhi = pd.read_csv("qualifying_info.csv",encoding = "unicode_escape" )
df_merc_21_22 = pd.read_csv("individual_performance.csv",encoding = "unicode_escape" )
df_fastlap_21_22_abudabhi = pd.read_csv("fastest_laps_info.csv",encoding = "unicode_escape" )
df_max_ver_21_22_eachrace = pd.read_csv("Driver_standings_of_driver.csv",encoding = "unicode_escape" )
df_driver_std_21_22 = pd.read_csv("driver_standings_all.csv",encoding = "unicode_escape" )
df_const_std_21_22 = pd.read_csv("constructor_standings.csv",encoding = "unicode_escape" )
```

```python
df_abudabhi = pd.merge(df_quali_21_22_abudabhi,df_fastlap_21_22_abudabhi, how="inner",on="DRIVER STANDINGS")
df_abudabhi
```

```python
points_team = df_merc_21_22.groupby(['GRAND PRIX',"YEAR"], as_index=False).sum().sort_values(by="PTS",ascending=False)
sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(data=points_team, x='GRAND PRIX',y='PTS',hue='YEAR')
plt.xticks(rotation = 70)

```

```python
# this chart shows the comparision of points scored by merceded in 2021 vs 2022 on the same track
```

```python
# the same thing can be viewed with a scatter plot 
sns.scatterplot(x='GRAND PRIX',y='PTS',hue = "YEAR",data=df_merc_21_22,size="PTS")
sns.set(rc={'figure.figsize':(20,20)})
plt.xticks(rotation = 70)
# this is a scatter plot
```
